```python
import nltk
from nltk.corpus import reuters
from collections import defaultdict
import random
import ipywidgets as widgets
from IPython.display import display, HTML

# Tokenize the text from Reuters corpus
words = nltk.word_tokenize(' '.join(reuters.words()))

# Function to generate n-grams
def generate_ngrams(words, n):
    return list(nltk.ngrams(words, n))

# Function to build an n-gram model
def build_ngram_model(words, n):
    ngrams = generate_ngrams(words, n)
    model = defaultdict(lambda: defaultdict(int))

    # Count frequency of n-grams
    for ngram in ngrams:
        prefix = ngram[:-1]  # The first n-1 words
        suffix = ngram[-1]  # The last word
        model[prefix][suffix] += 1

    # Convert counts to probabilities
    for prefix in model:
        total_count = float(sum(model[prefix].values()))
        for suffix in model[prefix]:
            model[prefix][suffix] /= total_count

    return model

# Function to predict multiple next words to allow user to choose
def predict_multiple_words(model, prefix, num_suggestions=3):
    next_words = model[prefix]
    if next_words:
        sorted_words = sorted(next_words.items(), key=lambda x: x[1], reverse=True)
        return [word for word, _ in sorted_words[:num_suggestions]]
    else:
        return []

# Global variable to store the current suggestions and the last typed words
current_suggestions = []
last_context = []

# Function to update suggestions based on typed text
def update_suggestions(change):
    global current_suggestions, last_context
    text = text_area.value  # Get the text typed so far
    words_in_text = text.split()

    # If there are at least n-1 words, use them to predict the next word
    if len(words_in_text) >= n-1:
        last_context = words_in_text[-(n-1):]  # Update the context
        prefix = tuple(last_context)
        current_suggestions = predict_multiple_words(ngram_model, prefix)
    else:
        current_suggestions = []

    # Update the suggestion display
    if current_suggestions:
        suggestions_text = " | ".join([f"<span style='color:gray;'>{s}</span>" for s in current_suggestions])
        overlay.value = f"{text} {suggestions_text}"
    else:
        overlay.value = text  # No suggestions, just show the typed text

# Functions to insert specific suggestions and shift the context
def insert_first_suggestion(change):
    global current_suggestions, last_context
    if current_suggestions:
        selected_suggestion = current_suggestions[0]  # Insert the first suggestion
        text_area.value += " " + selected_suggestion
        # Update context after inserting the suggestion
        last_context.append(selected_suggestion)
        last_context = last_context[1:]  # Keep only the last n-1 words
        update_suggestions(None)  # Refresh suggestions after inserting

def insert_second_suggestion(change):
    global current_suggestions, last_context
    if len(current_suggestions) >= 2:
        selected_suggestion = current_suggestions[1]  # Insert the second suggestion
        text_area.value += " " + selected_suggestion
        # Update context after inserting the suggestion
        last_context.append(selected_suggestion)
        last_context = last_context[1:]  # Keep only the last n-1 words
        update_suggestions(None)  # Refresh suggestions after inserting

def insert_third_suggestion(change):
    global current_suggestions, last_context
    if len(current_suggestions) >= 3:
        selected_suggestion = current_suggestions[2]  # Insert the third suggestion
        text_area.value += " " + selected_suggestion
        # Update context after inserting the suggestion
        last_context.append(selected_suggestion)
        last_context = last_context[1:]  # Keep only the last n-1 words
        update_suggestions(None)  # Refresh suggestions after inserting

# Create widgets
text_area = widgets.Textarea(
    description='Text:',
    placeholder='Start typing...',
    layout=widgets.Layout(width='100%', height='100px')
)
text_area.observe(update_suggestions, names='value')

# Create the suggestion box
overlay = widgets.HTML(value="", layout=widgets.Layout(width='100%'))

# Add the buttons to insert specific suggestions
first_suggestion_button = widgets.Button(description="Insert 1st Suggestion")
first_suggestion_button.on_click(insert_first_suggestion)

second_suggestion_button = widgets.Button(description="Insert 2nd Suggestion")
second_suggestion_button.on_click(insert_second_suggestion)

third_suggestion_button = widgets.Button(description="Insert 3rd Suggestion")
third_suggestion_button.on_click(insert_third_suggestion)

# Allow the user to select the n-gram size
n_selector = widgets.IntSlider(
    value=3,
    min=1,
    max=5,
    step=1,
    description="N-gram size:",
    continuous_update=False
)

# Update n-gram model when the n-gram size changes
def update_ngram_model(change):
    global n, ngram_model
    n = change['new']
    ngram_model = build_ngram_model(words, n)
    update_suggestions(None)  # Refresh suggestions after model change

n_selector.observe(update_ngram_model, names='value')

# Display the widgets
display(widgets.VBox([n_selector, text_area, overlay, widgets.HBox([first_suggestion_button, second_suggestion_button, third_suggestion_button])]))

# Build an initial trigram model
n = 3  # Default to trigram model
ngram_model = build_ngram_model(words, n)
```
